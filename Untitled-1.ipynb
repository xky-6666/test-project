```python
import pandas as pd
import numpy as np
import os

# ===================== 1. 读取解压后的txt文件 =====================
txt_path = "./data_raw/TCGA.BRCA.sampleMap_HiSeqV2.txt"

df = pd.read_csv(txt_path, sep="\t", index_col=0)
print(f"读取成功！原始数据维度（基因数 x 样本数）：{df.shape}")

# ===================== 2. 清洗1：去除低表达基因 =====================
min_exp = 1
min_sample = 0.5 * df.shape[1]
keep_gene = (df > min_exp).sum(axis=1) >= min_sample
df_filtered = df.loc[keep_gene, :].copy()

print(f"\n过滤低表达基因后维度：{df_filtered.shape}")

# ===================== 3. 清洗2：构建样本分组信息（关键修正） =====================
def get_sample_type(sample_id):
    try:
        code = sample_id.split("-")[3][:2]
        if code in ["01", "02", "03", "04", "05", "06", "07", "08", "09"]:
            return "Tumor"
        elif code in ["10", "11", "12", "13", "14"]:
            return "Normal"
        else:
            return "Other"
    except:
        return "Other"

# 先按df_filtered的列顺序生成分组信息，保证顺序一致
sample_ids = df_filtered.columns.tolist()
sample_info = pd.DataFrame({
    "sample_id": sample_ids,
    "group": [get_sample_type(s) for s in sample_ids]
})

# 检查一下分组统计
print("\n所有样本分组统计：")
print(sample_info["group"].value_counts())

# 只保留肿瘤和正常样本
keep_mask = sample_info["group"].isin(["Tumor", "Normal"])
df_clean = df_filtered.loc[:, keep_mask.values]
sample_info_clean = sample_info[keep_mask].copy()

print(f"\n保留肿瘤+正常样本后维度：{df_clean.shape}")
print("清洗后样本分组统计：")
print(sample_info_clean["group"].value_counts())

# ===================== 4. 保存清洗后的数据 =====================
output_dir = "./data_processed"
os.makedirs(output_dir, exist_ok=True)

clean_expr_path = os.path.join(output_dir, "TCGA-BRCA_clean_expression.csv")
clean_group_path = os.path.join(output_dir, "TCGA-BRCA_sample_groups.csv")

df_clean.to_csv(clean_expr_path)
sample_info_clean.to_csv(clean_group_path, index=False)

print(f"\n 数据清洗完成！")
print(f"干净表达矩阵已保存为：{clean_expr_path}")
print(f"样本分组信息已保存为：{clean_group_path}")

# ===================== 5. 快速检查 =====================
print("\n前5个基因，前5个样本的表达矩阵：")
print(df_clean.iloc[:5, :5])
```

```python
# ==============================================
# 图1：PCA 图
# ==============================================
from matplotlib.patches import Ellipse
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from scipy.stats import chi2  # 用这个，不是 sklearn 的 chi2

# 1. 准备 PCA 数据
top_var_genes = expr.var(axis=1).sort_values(ascending=False).head(1000).index
expr_pca = expr.loc[top_var_genes]

# PCA 格式：行=样本，列=基因
X = expr_pca.T

# 标准化
X_scaled = StandardScaler().fit_transform(X)

# 执行PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)
pc1_var = pca.explained_variance_ratio_[0] * 100
pc2_var = pca.explained_variance_ratio_[1] * 100

pca_df = pd.DataFrame({
    "sample_id": X.index,
    "PC1": pca_result[:, 0],
    "PC2": pca_result[:, 1]
})

#合并分组信息
pca_df = pca_df.merge(sample_info, on="sample_id")

#2. 绘图风格
palette = {
    "Normal": "#1f77b4",
    "Tumor": "#ffa9ce"
}

fig, ax = plt.subplots(figsize=(7, 6))

#散点图
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="group",
    palette=palette,
    s=30,
    ax=ax
)

#95% 置信椭圆函数
def draw_ellipse(ax, points, color, alpha=0.15):
    if len(points) < 3:
        return
    cov = np.cov(points.T)
    mean = np.mean(points, axis=0)
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    vals = vals[order]
    vecs = vecs[:, order]

    theta = np.degrees(np.arctan2(vecs[1, 0], vecs[0, 0]))
    scale = np.sqrt(chi2.ppf(0.95, df=2))
    width = 2 * scale * np.sqrt(vals[0])
    height = 2 * scale * np.sqrt(vals[1])

    ellipse = Ellipse(
        xy=mean, width=width, height=height, angle=theta,
        facecolor=color, edgecolor=color, alpha=alpha, lw=1
    )
    ax.add_patch(ellipse)

#画椭圆
for group in pca_df["group"].unique():
    group_points = pca_df.loc[pca_df["group"] == group, ["PC1", "PC2"]].values
    draw_ellipse(ax, group_points, palette[group])

#添加十字参考线
ax.axhline(0, color="gray", linestyle="--", linewidth=0.8)
ax.axvline(0, color="gray", linestyle="--", linewidth=0.8)

ax.set_xlabel(f"PC1 ({pc1_var:.1f}%)", fontsize=12)
ax.set_ylabel(f"PC2 ({pc2_var:.1f}%)", fontsize=12)
ax.set_title("PCA Plot: Normal vs Tumor", fontsize=15, pad=15)

plt.xlim(-40, 40)
plt.ylim(-30, 30)

ax.legend(title="Group", bbox_to_anchor=(1.02, 0.8), loc="upper left")

plt.tight_layout()
plt.savefig('./figures/PCA.png', 
            dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
plt.savefig("./figures/PCA.png", dpi=300, bbox_inches="tight")
plt.close()
print("PCA 图已保存: ./figures/PCA.png")
```

```python
# ==============================================
#火山图
# ==============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
from statsmodels.stats.multitest import multipletests

# ===================== 自动加载清洗好的数据 =====================
expr = pd.read_csv("./data_processed/TCGA-BRCA_clean_expression.csv", index_col=0)
sample_info = pd.read_csv("./data_processed/TCGA-BRCA_sample_groups.csv")

# 1. 分组样本提取
control_cols = expr.columns[expr.columns.isin(sample_info.loc[sample_info["group"] == "Normal", "sample_id"])]
treat_cols   = expr.columns[expr.columns.isin(sample_info.loc[sample_info["group"] == "Tumor", "sample_id"])]

# 2. 遍历每个基因做差异分析
results = []
for gene in expr.index:
    control = expr.loc[gene, control_cols].values
    treat   = expr.loc[gene, treat_cols].values

    mean_control = np.mean(control)
    mean_treat   = np.mean(treat)
    log2fc       = mean_treat - mean_control

    # Welch t test
    t_stat, p_val = ttest_ind(treat, control, equal_var=False)
    results.append([gene, mean_control, mean_treat, log2fc, p_val])

# 3. 构建结果表
result_df = pd.DataFrame(results, columns=[
    "Gene", "Mean_Normal", "Mean_Tumor", "log2FC", "pvalue"
])

# 4. FDR 校正
result_df["FDR"] = multipletests(result_df["pvalue"], method="fdr_bh")[1]

# 5. Y轴：-log10(FDR)
result_df["neg_log10_FDR"] = -np.log10(result_df["FDR"] + 1e-300)

# 6. 阈值设定
fc_threshold = 1
fdr_threshold = 0.05
sig_threshold = -np.log10(fdr_threshold)

# 7. 基因分类
result_df["status"] = "Not Significant"
result_df["status"] = np.where(
    (result_df["log2FC"] > fc_threshold) & (result_df["FDR"] < fdr_threshold), "Up",
    np.where(
        (result_df["log2FC"] < -fc_threshold) & (result_df["FDR"] < fdr_threshold), "Down",
        "Not Significant"
    )
)

print(result_df["status"].value_counts())

# ---------------------- 绘图 ----------------------
plt.figure(figsize=(9, 7))
plt.rcParams["figure.dpi"] = 130

palette = {
    "Up": "#ff6b6b",
    "Down": "#4dabf7",
    "Not Significant": "#cccccc"
}
hue_order = ["Up", "Down", "Not Significant"]

sns.scatterplot(
    data=result_df,
    x="log2FC",
    y="neg_log10_FDR",
    hue="status",
    hue_order=hue_order,
    palette=palette,
    s=28,
    alpha=0.85,
    edgecolor=None
)

# 图例
ax = plt.gca()
handles, labels = ax.get_legend_handles_labels()
ax.legend(
    handles=handles,
    labels=["Up", "Down", "Not significant"],
    title="Genes",
    loc="center left",
    bbox_to_anchor=(1.02, 0.5),
    fontsize=11,
    title_fontsize=12,
    frameon=True
)

# 阈值线
plt.axvline(x=fc_threshold, color="black", linestyle="--", alpha=0.6)
plt.axvline(x=-fc_threshold, color="black", linestyle="--", alpha=0.6)
plt.axhline(y=sig_threshold, color="black", linestyle="--", alpha=0.6)

# 坐标轴
plt.title("Volcano Plot (Normal vs Tumor)", fontsize=16, pad=15)
plt.xlabel("log2(Fold Change)", fontsize=13)
plt.ylabel("-log10(FDR)", fontsize=13)

plt.tight_layout()
plt.savefig('./figures/heatmap_expression.png', 
            dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
plt.savefig("./figures/Volcano.png", dpi=300, bbox_inches="tight")
plt.close()
print("火山图已保存: ./figures/Volcano.png")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch
from matplotlib.colors import ListedColormap

# 设置绘图参数
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.dpi'] = 150

# 1. 读取数据
expr_df = pd.read_csv("./data_processed/TCGA-BRCA_clean_expression.csv", index_col=0)
sample_info = pd.read_csv("./data_processed/TCGA-BRCA_sample_groups.csv")

print(f"表达矩阵维度: {expr_df.shape}")

# 2. 按分组排序样本
sample_order = sample_info.sort_values('group')['sample_id'].tolist()
expr_sorted = expr_df[sample_order]
group_sorted = sample_info.set_index('sample_id').loc[sample_order]['group']

# 3. 选择高变异基因（Top 30）
gene_var = expr_sorted.var(axis=1)
top_n = 30
top_genes = gene_var.nlargest(top_n).index.tolist()
expr_heatmap = expr_sorted.loc[top_genes]

# 4. Z-score标准化
expr_zscore = expr_heatmap.sub(expr_heatmap.mean(axis=1), axis=0).div(expr_heatmap.std(axis=1), axis=0)

# 5. 绘制热图
fig = plt.figure(figsize=(14, 10))

# 调整网格布局：给左侧留出更多空间放基因名
gs = fig.add_gridspec(2, 2, 
                      width_ratios=[0.03, 20], 
                      height_ratios=[1, 0.03],
                      hspace=0.01, 
                      wspace=0.02,
                      left=0.12,   # 给Y轴基因名留出足够空间
                      right=0.9)

# 主热图区域
ax_heatmap = fig.add_subplot(gs[0, 1])

# 分组颜色条区域
ax_colors = fig.add_subplot(gs[0, 0])

# 颜色条区域
ax_cbar = fig.add_subplot(gs[1, 1])

# 绘制分组颜色条（使用数值编码）
group_numeric = group_sorted.map({'Normal': 0, 'Tumor': 1}).values.reshape(-1, 1)
group_cmap = ListedColormap(['#2E7D32', '#C62828'])

ax_colors.imshow(group_numeric, aspect='auto', cmap=group_cmap, vmin=0, vmax=1)
ax_colors.set_xticks([])
ax_colors.set_yticks([])
# 修正1：Samples标签放在分组色条顶部，不再挤在Y轴上
ax_colors.set_title('Samples', fontsize=10, pad=5)

# 绘制热图
sns.heatmap(
    expr_zscore,
    ax=ax_heatmap,
    cmap='RdBu_r',
    center=0,
    vmin=-2.5,
    vmax=2.5,
    cbar_ax=ax_cbar,
    cbar_kws={'orientation': 'horizontal', 'label': 'Z-score'},
    xticklabels=False,
    yticklabels=True,
    linewidths=0
)

# 设置y轴标签：右对齐，避免被轴框遮挡
ax_heatmap.set_yticklabels(ax_heatmap.get_yticklabels(), fontsize=8, ha='right')
ax_heatmap.set_xlabel('Samples', fontsize=11)
ax_heatmap.set_ylabel('Genes', fontsize=11)

# 添加分组边界线
normal_count = (group_sorted == 'Normal').sum()
ax_heatmap.axvline(x=normal_count, color='white', linewidth=2.5, linestyle='-')
ax_colors.axvline(x=0.5, color='white', linewidth=2.5, linestyle='-')

# 添加图例
legend_elements = [
    Patch(facecolor='#2E7D32', label=f'Normal (n={normal_count})'),
    Patch(facecolor='#C62828', label=f'Tumor (n={len(group_sorted)-normal_count})')
]
ax_heatmap.legend(
    handles=legend_elements,
    loc='upper right',
    bbox_to_anchor=(1.15, 1.02),
    fontsize=9,
    frameon=True,
    fancybox=True
)

# 标题
fig.suptitle('TCGA-BRCA Gene Expression Heatmap\nTop 30 High-Variance Genes', 
             fontsize=13, fontweight='bold', y=0.98)

plt.savefig('./figures/heatmap_expression.png', 
            dpi=300, bbox_inches='tight', facecolor='white', edgecolor='none')
print("\n热图已保存至: ./figures/heatmap_expression.png")
plt.show()
```

火山图和热图两张图筛选出的基因不一样，是因为它们用了两种完全不同的筛选逻辑。

1. 火山图的筛选逻辑：“差异显著性”
火山图筛选的是差异表达基因（DEGs），它的核心是看基因在两组之间的表达差异是否显著。
计算标准：同时满足 |log2FC| > 1 和 FDR < 0.05
代表的信息：基因在肿瘤和正常组织中的表达水平是否有统计学意义上的巨大差异。
特点：只关注 “组间差异”，不管基因本身的表达量高低。

2. 热图的筛选逻辑：“表达变异度”
热图筛选的是高变异基因（High-variance genes），它的核心是看基因在所有样本中的表达波动是否大。
计算标准：计算每个基因在所有样本中的表达量方差（Variance），取 Top 30。
代表的信息：基因在整个数据集中的表达是否稳定。
特点：只关注 “整体波动”，不管这种波动是来自组间差异还是组内差异。

为什么结果会不一样？

举个例子：
基因 A：在正常组织中表达很高，在肿瘤中表达极低，差异非常显著（会出现在火山图里）。但它在正常组内表达很稳定，肿瘤组内也很稳定，整体方差不大（可能不会出现在热图里）。
基因 B：在正常和肿瘤组织中的平均表达水平差不多，所以差异不显著（不会出现在火山图里）。但它在部分样本中表达极高，部分极低，整体波动很大（会出现在热图里）。

简单来说：
火山图筛选的是 “组间差异大” 的基因。
热图筛选的是 “整体波动大” 的基因。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.dpi'] = 150

# 1. 读数据
expr_df = pd.read_csv("./data_processed/TCGA-BRCA_clean_expression.csv", index_col=0)
sample_info = pd.read_csv("./data_processed/TCGA-BRCA_sample_groups.csv")

# 2. 转成 样本-基因 格式
expr_T = expr_df.T.reset_index().rename(columns={"index":"sample_id"})

# 3. 和分组合并
box_df = expr_T.merge(sample_info, on="sample_id")

# 4. 计算每个样本所有基因的中位表达量（整体表达水平）
box_df["median_expr"] = box_df.drop(["sample_id","group"], axis=1).median(axis=1)

# 5. 绘图
plt.figure(figsize=(5,6))
palette = {"Normal":"#1f77b4", "Tumor":"#ffa9ce"}

sns.boxplot(
    x="group",
    y="median_expr",
    data=box_df,
    palette=palette,
    width=0.6,
    fliersize=3
)

plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.title("Overall Expression Distribution", fontsize=13)
plt.xlabel("Group", fontsize=11)
plt.ylabel("Median Expression Level", fontsize=11)
plt.tight_layout()

from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor=palette["Normal"], label='Normal'),
    Patch(facecolor=palette["Tumor"], label='Tumor')
]
plt.legend(handles=legend_elements, title="Group", 
           bbox_to_anchor=(1.02, 1), loc="upper left")

plt.savefig("./figures/boxplot_overall_expression.png", 
            dpi=300, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.dpi'] = 150

# 1. 读取数据
expr_df = pd.read_csv("./data_processed/TCGA-BRCA_clean_expression.csv", index_col=0)
sample_info = pd.read_csv("./data_processed/TCGA-BRCA_sample_groups.csv")

# 2. 整理数据：每个样本的整体表达中位数
sample_expr = expr_df.median(axis=0).reset_index()
sample_expr.columns = ["sample_id", "expression"]
plot_df = sample_expr.merge(sample_info, on="sample_id")

# 3. 统计检验：Tumor vs Normal
tumor_expr = plot_df[plot_df["group"] == "Tumor"]["expression"]
normal_expr = plot_df[plot_df["group"] == "Normal"]["expression"]
t_stat, p_val = ttest_ind(tumor_expr, normal_expr, equal_var=False)
p_text = f"P = {p_val:.2e}" if p_val < 0.001 else f"P = {p_val:.3f}"

# 4. 绘图
plt.figure(figsize=(5, 6))

# 配色（和你的热图/PCA图统一）
palette = {
    "Tumor": "#f94949",
    "Normal": "#449ee7"
}

# 小提琴图
sns.violinplot(
    data=plot_df,
    x="group",
    y="expression",
    palette=palette,
    linewidth=1.2,
    inner="quartile", # 内部显示四分位数线
)

# 添加散点（不重叠，更清晰）
sns.stripplot(
    data=plot_df,
    x="group",
    y="expression",
    color="white",
    edgecolor="black",
    linewidth=0.3,
    size=3,
    jitter=0.2,
    alpha=0.8
)

# 添加显著性标记
max_y = plot_df["expression"].max()
plt.plot([0, 1], [max_y+0.02, max_y+0.02], color="black", lw=1)
#P < 0.0001 用 **** 表示
plt.text(0.5, max_y+0.04, "****", ha="center", fontsize=12)

# 美化坐标轴
plt.title("Overall Expression Distribution", fontsize=13, pad=15)
plt.xlabel("Group", fontsize=11)
plt.ylabel("Median log2 Expression", fontsize=11)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.ylim(bottom=plot_df["expression"].min() - 0.1)

plt.tight_layout()
plt.savefig("./figures/violin_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.dpi'] = 150

# 读取数据
expr_df = pd.read_csv("./data_processed/TCGA-BRCA_clean_expression.csv", index_col=0)
sample_info = pd.read_csv("./data_processed/TCGA-BRCA_sample_groups.csv")

# ===================== 关键：计算每个样本的中位数=====================
sample_expr = expr_df.median(axis=0).reset_index()
#对每个样本，计算这个样本里 “所有基因的中位数”--每个样本只保留 1 个数字
#总共 1200 个样本 → 1200 个数字
sample_expr.columns = ["sample_id", "expression"]
plot_df = sample_expr.merge(sample_info, on="sample_id")

# ===================== 绘图（恢复你原来的分布！）=====================
plt.figure(figsize=(7, 5))

# 先画蓝色（Normal），再画红色（Tumor）
sns.kdeplot(
    data=plot_df[plot_df["group"] == "Normal"],
    x="expression",
    color="#4dabf7",
    fill=True,
    alpha=0.7,
    linewidth=2,
    label="Normal"
)

#画红色（Tumor）的密度曲线
#数据：1104 个肿瘤样本的表达中位数
sns.kdeplot(
    data=plot_df[plot_df["group"] == "Tumor"],
    x="expression",
    color="#ff6b6b",
    fill=True,
    alpha=0.5,
    linewidth=2,
    label="Tumor"
)

plt.xlabel("Median Expression Level (log2)", fontsize=11)
plt.ylabel("Density", fontsize=11)
plt.title("Expression Distribution of Normal vs Tumor Samples", fontsize=13)
plt.legend()
plt.grid(alpha=0.2, linestyle="--")
plt.tight_layout()
plt.savefig("./figures/density_plot_original.png", dpi=300, bbox_inches="tight")
plt.show()
```
